# Applying Elbow Technique

```python
k_range = range (1,10)
sse=[]

for k in k_range:
    k_mean = KMeans(n_clusters = k)
    k_mean.fit(iris_dataset[[ 'petal_length', 'petal_width']])
    sse.append(k_mean.inertia_)
```

```python
sse
```

```python
plt.xlabel('k_range')
plt.ylabel('Sum of Squared errors')
plt.plot(k_range,sse)
```

# Applying KMeans Algorithm

```python
k_mean = KMeans(n_clusters=3, random_state=0,)
Y_pred = k_mean.fit_predict(iris_dataset[[ 'petal_length', 'petal_width']])
Y_pred
```

```python
iris_dataset['cluster'] = Y_pred
iris_dataset
```

# Ploting confusion matrix

```python
from sklearn.metrics import confusion_matrix 
conf_mat = confusion_matrix(iris_dataset.species, iris_dataset.cluster)
conf_mat
```

```python
true_labels = iris_dataset.species 
predicted_labels = iris_dataset.cluster

conf_mat = confusion_matrix(true_labels, predicted_labels)
class_labels = ['Setosa', 'Versicolor', 'Virginica']

plt.imshow(conf_mat, interpolation='nearest', cmap=plt.cm.Oranges)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(class_labels))

plt.xticks(tick_marks, class_labels)
plt.yticks (tick_marks, class_labels)

#Fill matrix with values for i in range(len(class_labels)):

for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j, i, str(conf_mat[i][j]), ha='center', va='center', color="white")

plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show
```
